```python
import copy

class Solution:
    def findMaxForm(self, strs, m, n):
        numOne = 0
        numZero = 0
        strList = copy.deepcopy(strs)
        addDict = []
        removeDict = []
        

        def lenghtValue(list_One_Zero):
            list = []
            for value in list_One_Zero:
                list.append(len(value))

            print("list",list)
            return list
        
        listLenghtValue = lenghtValue(strList)

        
        while True:

            print ("infinity")

            for element in strs:

                if not listLenghtValue:
                    break
                
                smallest = min(listLenghtValue)

                print ("len element",len(element))
                print ("smallest", smallest)

                if not element in addDict and not element in removeDict \
                and len(element) <= smallest:

                    for number in element:
                        print ('number treat', element)
                        print(number)
                        if numZero < m:
                            if number == "0":
                                numZero += 1
                        if numOne < n:
                            if number == "1":
                                numZero += 1

                        if numOne == m and numZero == n:
                            print("we break")
                            break

                    listLenghtValue.remove(smallest)
                    
                    strList.remove(element)
                    addDict.append(element)


            
        
s1 = Solution()
print (s1.findMaxForm(["10","0001","111001","1","0"], 5, 3))
```
